## 1. Funci√≥n para leer tablas

````python
# Creando funcion para leer las tablas de la base de datos
def leer_tabla(tabla, engine):
    query = "SELECT * FROM " + tabla
    df = pd.read_sql(query, engine)
    return df
````

## Leyendo tablas
Las tablas a leer son:

1. orders
2. orderdetails
3. customers
4. products
5. employees

````python
# Leyendo las tablas con leer_tabla
lista_tablas = ["orders", "orderdetails", "customers", "products", "employees"]
dfs = [leer_tabla(i, engine) for i in lista_tablas]
orders, orderdetails, customers, products, employees = dfs
````

## 2. Cruzando las tablas

````python
df = orders.merge(orderdetails, how= 'inner', on='orderNumber', validate= 'one_to_many')
df = df.merge(customers, how= 'inner', on='customerNumber', validate= 'many_to_one')
df = df.merge(products, how= 'inner', on='productCode', validate= 'many_to_one')
df = df.merge(employees, how= 'inner', left_on='salesRepEmployeeNumber', right_on='employeeNumber',validate= 'many_to_one')
df.head()
````

## 3. Agregando columnas

````python
df['venta'] = df['quantityOrdered']*df['priceEach']
df['costo'] = df['quantityOrdered']*df['buyPrice']
df['ganancia'] = df['venta']-df['costo']
df.head()
````

````python
df.columns
````

## 4. Total de ventas por l√≠nea de productos

````python
# Total de ventas por linea de producto
totales_linea_producto = pd.pivot_table(
    data = df
    , index = 'productLine'
    , values = 'venta'
    , aggfunc= 'sum'
    , margins = True # filas y columnas de totales
    , margins_name= 'total'
)
totales_linea_producto
````

import pandas as pd
# Creando funcion de filtro de fechas
def filtro_fechas(df, columna, fecha_inicio, fecha_fin):
    df[columna] = pd.to_datetime(df[columna])
    df_filtrado = df[df[columna].between(fecha_inicio, fecha_fin)]
    return df_filtrado

# Funcion funcion_agrupadora
def funcion_agrupadora(df, fila, valor, medida,orden):
    df_pivot = pd.pivot_table(
        data = df
        , index = fila
        #, columns = columna
        , values = valor
        , aggfunc= medida
        , margins = True # filas y columnas de totales
        , margins_name= 'total'
        , fill_value = 0 # se cambian los valores NaN por 0
    ).sort_values(by=orden, ascending=False).head(11) # ordena, toma total y los 10 primeros valores
    return df_pivot.sort_values(by=orden, ascending=True)

# Funcion para guardar df
